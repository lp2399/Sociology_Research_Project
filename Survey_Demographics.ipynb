<h1>Survey Questions Results Demographics</h1>
<h3>This section covers survey results from question 1 to 5</h3>

```python
import pandas as pd
%matplotlib inline
```

```python
dataFrame = pd.read_excel('Survey_Data_Demographics.xlsx')
dataFrame["Birthdate"]= pd.to_datetime(dataFrame["Birthdate"])
dataFrame['Age'] = pd.to_datetime('now').year-dataFrame['Birthdate'].dt.year
```

<h3>Participants were asked (Question 1) "What is your undergraduate program?"</h3>

```python
dataFrame["Undergraduate program"].value_counts().plot.bar() # Below is a bar graph from Question 1
```
